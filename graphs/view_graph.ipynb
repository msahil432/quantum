```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use(['seaborn'])
df=pd.read_excel("dff.xlsx")
df["Number"]=df["First Prime"]*df["Second Prime"]
df["Quantum Time"]=df["Programming Time"]+df["Sampling Time"]/df["Total Samples"]
df.sort_values("Number",inplace=True)
df
```

```python
fig=plt.figure(figsize=(15,10))
plt.plot(df.Number,df["Single Run"])
plt.plot(df.Number,df["Quantum Time"])
plt.xlabel("Number for Factorization",fontsize="large")
plt.ylabel("Time in micro seconds",fontsize="large")
plt.legend(labels=["Calssical Time","Quantum Time"],loc="upper right",fontsize="medium")
plt.title("Classical VS Quantum Time Comparison")
```

```python
dff=df.copy()
dff["Quantum Time"]=dff["Quantum Time"]/1000000                           #micro to seconds conversion
dff["Single Run"]=dff["Single Run"]/1000000
dff["For S Times"]=dff["For S Times"]/1000000
dff
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(dff.Number,dff["Single Run"])
plt.plot(dff.Number,dff["Quantum Time"])
plt.xlabel("Number for Factorization",fontsize="large")
plt.ylabel("Time in seconds",fontsize="large")
plt.legend(labels=["Calssical Time","Quantum Time"],loc="lower right",fontsize="medium")
plt.title("Classical VS Quantum Time Comparison")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.spines['left'].set_visible(False)
#ax.spines['bottom'].set_visible(False)
```
